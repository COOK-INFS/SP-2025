```python
import pandas as pd

dfTitles = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/titleBasics.tsv', sep='\t', low_memory=False)
dfRatings = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/ratings.tsv', sep='\t', low_memory=False)
dfNames = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/nameBasics.tsv', sep='\t', low_memory=False)
dfCrew = pd.read_csv('http://128.198.162.191/dataFiles/IMDB/crew.tsv', sep='\t', low_memory=False)
```

## Question 1: Movie Director with the Highest average rating and the most votes.


```python
# Look at the data to see what fields we need.
dfNames.head()
```

### Merge Ratings and Crew
This gets the director unique identifiers, averageRating, and numVotes into the same dataframe.

```python
ratingsAndCrew = pd.merge(dfRatings, dfCrew, on='tconst')

ratingsAndCrew
```

### Rename the Directors column to keep things straight for me and allow a merge.

```python
dfCrewRename = ratingsAndCrew.rename(columns={'directors': 'nconst'})

dfCrewRename
```

### Merge dfCrewRename and dfNames
Gets the names aligned with the nconst unique identifier

```python
ratingsCrewDir = pd.merge(dfCrewRename, dfNames, on='nconst')

ratingsCrewDir
```

### Sort the information 
This allows us to see the highest rating and number of votes.

```python
ratingsCrewDir.sort_values(by=['averageRating', 'numVotes'], ascending=False)
```

### Output the information

```python
semi1 = (ratingsCrewDir.sort_values(by=['averageRating', 'numVotes'], ascending=False))

# Select columns based on specification
semi2 = semi1[['primaryName', 'averageRating', 'numVotes']]

final = (semi2.head(1))

final
```

### Output results to file

```python
# Removed the index to meet the requirement specification.
final.to_csv('question1.csv', index=False)
```

# Question 2: Highest rated film for the year 2000 by votes then rating.
In this example I show how your code might look if you performed each action within a single code cell.

```python
# Merge dataFrames to get titles and ratings in the same dataFrame.
highRated = pd.merge(dfTitles, dfRatings, on='tconst')

# Query to remove all records except "movies" in the year 2000.
movieFilter = (highRated.query('titleType == "movie" & startYear == "2000"'))

# Sort the values to find the highest rated.
movieFilter.sort_values(by=['numVotes', 'averageRating'], ascending=False)

# Format the results for output
semi1 = (movieFilter.sort_values(by=['numVotes', 'averageRating'], ascending=False))
semi2 = semi1[['startYear', 'primaryTitle', 'numVotes', 'averageRating']]
final=(semi2.head(5))

# show final results to verify
final
```

```python
# Write the results to a file.
final.to_csv('question2.csv', index=False)
```
