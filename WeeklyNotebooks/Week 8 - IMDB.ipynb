# IMDB question

What was the name and year of the movie with the highest rating and highest number of votes?

```python
# Merge our dataFrames to get the info we want.

mvRating = pd.merge(dfTitles, dfRatings, on='tconst')

mvRating
```

```python
# Reduce the data to what we want to see (Movie).

ratingSort = (mvRating.query('titleType == "movie"'))

ratingSort
```

```python
# Sort the data to determine the ratings and votes.
ratingSort.sort_values(by='averageRating', ascending=False)

```

```python
# Sort on multiple columns

ratingSort.sort_values(by=['averageRating', 'numVotes'], ascending=False)
```

```python
# Sort on multiple columns

ratingSort.sort_values(by=['numVotes', 'averageRating'], ascending=False)
```

# Work the data to present to the boss.
What was the name and year of the movie with the highest rating and highest number of votes?

```python
# Create a dataframe with the queried and sorted data
semi1 = (ratingSort.sort_values(by=['numVotes', 'averageRating'], ascending=False))

# Create a dataframe with only the required columns
semi2 = semi1[['primaryTitle', 'startYear', 'numVotes', 'averageRating']]

# Select a subset of the data
final = (semi2.head(10))

# Output the results 
print(final)
```

```python
# Output the results to a file

final.to_csv('dataFiles/finalResults.csv', index=False)
```
