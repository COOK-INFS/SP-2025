```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2023-051124.txt')

df.head(15)
```

## Create column names for our dataframe

```python
file_path = "https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2023-051124.txt"

# Define our custom column names
column_names = [
   'ID-Date', 'Time', 'RecordID', 'Status', 'Latitude', 'Longitude',
   'WindSpeed', 'Pressure', 'ExtraField1', 'ExtraField2', 'ExtraField3',
   'ExtraField4', 'ExtraField5', 'ExtraField6', 'ExtraField7', 'ExtraField8',
   'ExtraField9', 'ExtraField10', 'ExtraField11', 'ExtraField12' 
]

# Apply the list to my dataframe
df = pd.read_csv(file_path, header=0, names=column_names, usecols=range(len(column_names)), on_bad_lines='skip')

df.head()
```

## Extract 'ID', 'Year' and 'Month' from the ID-Date column.

```python
# Create a Year and Month column with empty values
df['Year'] = ''
df['Month'] = ''

# Extract Year and Month from ID-Date
df.loc[~df['ID-Date'].str.contains('AL'), 'Year'] = df['ID-Date'].str[:4]
df.loc[~df['ID-Date'].str.contains('AL'), 'Month'] = df['ID-Date'].str[4:6]

# Create a new 'Unique-ID' column that copies the 'AL' identifier
df['Unique-ID'] = df['ID-Date'].apply(lambda x: x if 'AL' in x else None)

# Forward-fill the 'Unique-ID' column to repeat the 'AL' identifier
df['Unique-ID'] = df['Unique-ID'].ffill()

df.head()
```

## Total Hurricanes per year

```python
hurricanes_by_year = df.groupby('Year')['Unique-ID'].count().reset_index(name='COUNT')

hurricanes_by_year
```

### Plot the results
Keep aware of the '0' values in 'Year'

```python
# Convert the Year column to numeric, coercing the errors to NaN
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Drop the rows where 'Year' is NaN
df_clean = df.dropna(subset=['Year'])

# Convert our Year to an integer
df_clean.loc[:,'Year'] = df_clean['Year'].astype(int) # The ":'Year'" instructs Pandas to select all rows in 'Year'

# Group by the cleaned 'Year' data
hurricanes_by_year = df_clean.groupby('Year')['Unique-ID'].nunique()

# Calculate the cumulative average
cumulative_average = hurricanes_by_year.expanding().mean()

# Plot results
plt.figure(figsize =(10,5))
plt.plot(hurricanes_by_year.index, hurricanes_by_year.values, marker='o')

# Plot the cumulative average as a line
plt.plot(cumulative_average.index, cumulative_average.values, color='red', linestyle='--', label='Cumulative')

plt.xlabel('Year')
plt.ylabel('Number of Unique Hurricanes')
plt.title("Number of Unique Hurricanes per year")

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(10))
plt.xticks(rotation=45)

plt.grid(True)
plt.legend()
plt.show
```

## Total Hurricanes by Category and Year

```python
# Create a function to use wind speeds to calculate category
def categorize_wind_speed(speed):
    if 74 <= speed <= 95:
        return "Category 1"
    elif 96 <= speed <= 110:
        return 'Category 2'
    elif 111 <= speed <= 129:
        return 'Category 3'
    elif 130 <= speed <= 156:
        return 'Category 4'
    elif speed >= 157:
        return 'Category 5'
    

df['Category'] = df['WindSpeed'].apply(categorize_wind_speed)

hurricanes_by_year_category = df.groupby(['Year', 'Category'])['Unique-ID'].count().unstack(fill_value=0)

plt.figure(figsize=(12, 6))

for category in hurricanes_by_year_category.columns:
    plt.plot(hurricanes_by_year_category.index, hurricanes_by_year_category[category], label=category)

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Number of Hurricanes by category and Year')
plt.grid()
plt.legend()
plt.show()
```

```python
# First, we need to install pandas if you haven't already


import pandas as pd

# Load the HURDAT2 dataset
file_path = "https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2023-051124.txt"  # Replace this with the path to your HURDAT2 data file

# Read the HURDAT2 file into a dataframe
hurdat2 = pd.read_csv(file_path, header=None, names=['ID', 'Date', 'Time', 'Status', 'Latitude', 'Longitude', 'Wind', 'Pressure', 'Additional Info'], sep='|')

# Display the first few rows of the dataset to inspect it
hurdat2.head()

# Now we need to filter out the data for hurricanes (status "HU" means Hurricane)
hurdat2['Year'] = pd.to_datetime(hurdat2['Date'], format='%Y%m%d').dt.year  # Extract the year from the 'Date' column

# Filter for Hurricane (HU) status only
hurricanes = hurdat2[hurdat2['Status'] == 'HU']

# Group by year and count the number of hurricanes each year
hurricanes_by_year = hurricanes.groupby('Year').size()

# Display the number of hurricanes by year
hurricanes_by_year

# Plotting the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
hurricanes_by_year.plot(kind='bar', color='skyblue')
plt.title('Number of Hurricanes per Year')
plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
